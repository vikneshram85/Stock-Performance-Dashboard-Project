```python
import pandas as pd
from sqlalchemy import create_engine

# 1️⃣ Read CSV
csv_file = r"C:\Users\Nidish Kumaar V\OneDrive\Viknesh\GUVI - Data Science Course Materials\Capstone Projects\Stock Performance Dashboard Project - Updated\Sector_data - Sheet1.csv"
df = pd.read_csv(csv_file)

# 2️⃣ Create MySQL connection
# Format: mysql+mysqlconnector://username:password@host:port/database
engine = create_engine("mysql+mysqlconnector://root:root@localhost:3306/stock_data_db")

# 3️⃣ Load CSV into MySQL
df.to_sql(
    name='sector_data',
    con=engine,
    if_exists='replace',  # use 'append' to add rows
    index=False
)

print("CSV data loaded into MySQL successfully!")
```
